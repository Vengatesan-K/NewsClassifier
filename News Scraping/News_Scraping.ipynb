```python
url = 'https://timesofindia.indiatimes.com/briefs'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find all <h2> elements as headings
headings = soup.find_all('h2')

# Find all <p> elements as content
contents = soup.find_all('p')

# Extracting text from headings and contents
headings_text = [heading.get_text().strip() for heading in headings]
contents_text = [content.get_text().strip() for content in contents]

# Zipping headings and contents together
data = zip(headings_text, contents_text)

# Specify the CSV file name
csv_filename = 'headings_contents.csv'

# Writing data to CSV file
with open(csv_filename, 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    # Write header row
    writer.writerow(['Heading', 'Content'])
    # Write rows of heading and content pairs
    writer.writerows(data)

print(f"Data has been saved to '{csv_filename}'.")
```

```python
url = 'https://timesofindia.indiatimes.com/briefs'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find all <h2> elements as headings
headings = soup.find_all('h2')

# Create a dictionary to store heading-content pairs
data_dict = {}

# Loop through each heading and find its corresponding <p> content
for heading in headings:
    # Get the text of the heading
    heading_text = heading.get_text().strip()
    # Find the next <p> element after the current heading
    next_p = heading.find_next('p')
    if next_p:
        # Get the text of the content
        content_text = next_p.get_text().strip()
        # Store heading-content pair in the dictionary
        data_dict[heading_text] = content_text

# Specify the CSV file name
csv_filename = 'headings_contents1.csv'

# Writing data to CSV file
with open(csv_filename, 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    # Write header row
    writer.writerow(['Heading', 'Content'])
    # Write rows of heading and content pairs
    for heading, content in data_dict.items():
        writer.writerow([heading, content])

print(f"Data has been saved to '{csv_filename}'.")
```

```python
articles_data = []

for article_url in article_urls:
    response = requests.get(article_url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract and display heading and content
        heading = soup.find(class_="headline__text inline-placeholder")
        content = soup.find(class_="paragraph inline-placeholder")

        if heading and content:
            articles_data.append({
                "Heading": heading.text.strip(),
                "Content": content.text.strip()
            })
        else:
            articles_data.append({
                "Heading": "Heading not found",
                "Content": "Content not found"
            })

    else:
        articles_data.append({
            "Heading": "Failed to retrieve content",
            "Content": f"Failed URL: {article_url}"
        })

# Save extracted headings and content as a CSV file
csv_file = 'articles_data.csv'

with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    fieldnames = ['Heading', 'Content']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    
    writer.writeheader()
    for article in articles_data:
        writer.writerow(article)

print(f"Data saved to {csv_file}")
```

```python
# Function to scrape individual section of the website
def scrape_section_data(section_url):
    # Send a GET request to the section URL
    response = requests.get(section_url)
    section_data = []

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')
        articles = soup.find_all('div', class_='gs-c-promo')

        # Extract titles and content of each article in the section
        for article in articles:
            title_element = article.find('h3', class_='gs-c-promo-heading__title')
            title = title_element.text.strip() if title_element else 'No title'

            content_element = article.find('p', class_='gs-c-promo-summary')
            content = content_element.text.strip() if content_element else 'No content available'

            section_data.append({'Title': title, 'Content': content})

    else:
        print(f'Failed to fetch data from {section_url}')

    return section_data
```

```python
# Function to save data to CSV file
def save_to_csv(data, file_name='bbc_news_scrap.csv'):
    with open(file_name, 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['Title', 'Content']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        # Write rows
        for row in data:
            writer.writerow(row)

    print(f"Data saved to '{file_name}'")
```

```python
# Function to scrape the BBC News website
def scrape_bbc_news():
    base_url = 'https://www.bbc.com'
    url = 'https://www.bbc.com/news'
    data = []

    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')
        sections = soup.find_all('a', class_='nw-o-link')

        # Iterate through sections and scrape data
        for section in sections:
            section_name = section.text.strip()
            section_link = urljoin(base_url, section['href'])

            if section_link.startswith(base_url):
                print(f"Scraping articles from '{section_name}' section...")
                section_data = scrape_section_data(section_link)
                data.extend(section_data)
                print("\n")

        # Save data to CSV file
        save_to_csv(data)

    else:
        print('Failed to fetch data from BBC News')
```

```python
url = 'https://www.thehindu.com/latest-news/'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find all <h3> elements with class 'title' to extract article URLs
article_links = soup.find_all('h3', class_='title')

# Extracting href attribute values from article links
article_urls = [link.find('a').get('href') for link in article_links]

# Create a list to store headings and contents
data = []

# Function to extract heading and content from an article URL
def extract_data(article_url):
    article_response = requests.get(article_url)
    article_soup = BeautifulSoup(article_response.text, 'html.parser')
    
    # Find <h1> element as heading
    heading = article_soup.find('h1', class_='title').text.strip()
    
    # Find elements with class 'sub-title' as content
    content_elements = article_soup.find_all('h3', class_='sub-title')
    content = '\n'.join([elem.text.strip() for elem in content_elements])
    
    data.append((heading, content))

# Loop through each article URL and extract data
for article_url in article_urls:
    extract_data(article_url)

# Specify the CSV file name
csv_filename = 'headings_contents_thehindu.csv'

# Writing data to CSV file
with open(csv_filename, 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    # Write header row
    writer.writerow(['Heading', 'Content'])
    # Write rows of heading and content pairs
    writer.writerows(data)

print(f"Data has been saved to '{csv_filename}'.")
```
